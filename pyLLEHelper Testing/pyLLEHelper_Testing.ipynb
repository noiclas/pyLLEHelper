I edited the pyLLE import system. For the purposes of this demonstration, type 'H' when prompted by the pyLLE text.

```python
import pyLLEHelper as lle
import numpy as np
import matplotlib.pyplot as plt
```

As always, have to define res and sim to any sort of simulation. Using mode/resonance frequency data given in the dispfile seen in res. This is some real data provided by Mark

```python
res = dict(
        R=175e-6, 
        Qi=1e6, 
        Qc=1e6, 
        γ=3.2, 
        dispfile="comb300g_1600w_B_TM_long_polctrl_v2_06-15-2023_normedsweep_freqs.csv"
)

sim = dict(
    Pin=[600e-3], 
    f_pmp=[193e12],
    φ_pmp=[0], 
    δω=[None], 
    Tscan=0.2e6,
    μ_sim=[-50, 50],
    δω_init= 1e9 * 2 * np.pi,
    δω_end= -6.5e9 * 2 * np.pi,
    num_probe = 5000, 
)
```

pyLLEHelper shares the same initialization as LLESolver. Here is how to initialize it:

```python
p = lle.pyLLEHelper(sim=sim,res=res,debug=False)
p.plotDint()
```

We need to run a detuning sweep simulation in order to get soliton data. Once we have the soliton data we can mess with it.

use the detuningSweepSim() function to do this. saveSolver = True will save the pyLLEHelper "p" in a .pkl file in the current directory. The name of the .pkl is the data file name + the type of simulation ran (name will be printed once the simulation is complete)

```python
p.Analyze(True,plottype='sim')
```

```python
p.detuningSweepSim(saveSolver=True)
```

use plotIntraCavityPower() to do exactly what the function's name is.

```python
p.plotIntraCavityPower()
```

Ok, I want single soliton states so I'm gonna pick one on the last "stair step". This is roughly the range 2500-2850 of "LLE sub-sampled steps"

```python
p.PlotCombSpectra(2650)
```

Ok this looks like a decent comb at detuning step 2650. Seems like a good place to initialize a simulation from.

Good thing pyLLEHelper has freezeSoliton()! This will run the simulation from the initizlized state AND procceed to freeze it (i.e. correct the round trip time of the soliton)

		Params:
			deturningIdx - detuning index corresponding to wanted initial soliton state.
			saveSolvers - True if you want to save drifting and frozen solvers. 
                            Will save as .pkl in current dir.
			simDrifting - True if drifting sim is needed, False if not.
			plotDrifting - True to plot drifting plot, False to not plot.
			plotFrozen - True to plot frozen soliton plot, Flase to not plot.

		Returns:
			frozenSolver - pyLLEHelper object for frozen soliton.
			driftingSolver - pyLLEHelper object for drifting soliton.

```python
frozen, drifting = p.freezeSolitonSim(detuningIdx=2650,saveSolvers=True,simDrifting=True)
```

Now that we have a frozen soliton, whose data is held in "frozen", a pyLLEHelper, we can run some perturbations on it.

First up is detuning perturbation. In the following block, I am adding a little sine wave to the detuning to create my manual detuning array. It has be be length Tscan in order to work properly (for now at least)

```python
drifting.plotSolitonPos(),
frozen.plotSolitonPos()
```

```python
detuning = frozen.detuning
x = np.linspace(0,2*np.pi,int(frozen.sim.Tscan))
man_detun = (np.ones(int(frozen.sim.Tscan))+.11*np.sin(x))*detuning
```

Now its time to run a detuning perturbed simulation.

Use detuningPerturbationSim() as seen below.

		Params:
			Manual_Detuning - list/array of length Tsan. Holds the 
                perturbed detuning values for each round trip.
			plotSolitonPos - True if soliton position plot wanted, False if not.
			saveSolver - True if want to save pyLLEHelper as a .pkl.

		Returns:
			pinSolver - pyLLEHelper object with detuning perturbed data.
		

```python
detuningPerturbed = frozen.detuningPerturbationSim(Manual_Detuning=man_detun,saveSolver=True)
```

```python
detuningPerturbed.plotSolitonPos()
```

Wow! that soliton got pretty fucked up. We can plot the peak power in the cavity per time step to see what the amplitude is doing.

Use plotMaxAcavAmplitudes().

```python
detuningPerturbed.plotMaxAcavAmplitudes()
```

Next is input pump power (Pin) perturbation. For the manual Pin, I'm making the power go from 1x -> 6x -> 1x the actual Pin over the course of the simulation. 

```python
pin = frozen.Pin
x1 = np.append(np.linspace(0,5,int(frozen._sim["Tscan"]/2)),np.linspace(5,0,int(frozen._sim["Tscan"]/2)))
man_pin = (np.ones(int(frozen.sim.Tscan))+x1)*pin
```

Now its time to run a detuning perturbed simulation.

Use pinPerturbationSim() as seen below.

		Params:
			Manual_Pin - list/array of length Tsan. 
                Holds the perturbed input pump power for each round trip.
			plotSolitonPos - True if soliton position plot wanted, False if not.
			saveSolver - True if want to save pyLLEHelper as a .pkl.

		Returns:
			pinSolver - pyLLEHelper object with pin perturbed data.

```python
pinPerturbed = frozen.pinPerturbationSim(Manual_Pin=man_pin,saveSolver=True)
pinPerturbed.plotSolitonPos()
```

Again, we can plot what the peak of the soliton is doing over the course of the simulation

```python
pinPerturbed.plotMaxAcavAmplitudes()
```

Lets see what happens when detuning perturbation is added to the Pin perturbed simulation

```python
#This is probably non-physical as hell but it should at least do something
megaPerturbed0 = detuningPerturbed.pinPerturbationSim(Manual_Pin=man_pin,saveSolver=False)
```

```python
megaPerturbed0.plotSolitonPos()
```

Here are the peaks in the cavity

```python
megaPerturbed0.plotMaxAcavAmplitudes()
```

I even made a function that will run both of these perturbations simultaneously.

The function is detuningAndPinPerturbationSim()

		Params:
			Manual_Detuning - list/array of length Tsan. 
                Holds the perturbed detuning values for each round trip.
			Manual_Pin -  list/array of length Tsan. 
                Holds the perturbed input pump power for each round trip.
			plotSolitonPos - True if soliton position plot wanted, False if not.
			saveSolver - True if want to save pyLLEHelper as a .pkl.

		Returns:
			doublyPerturbedSolver - pyLLEHelper object with detuning and Pin perturbed data.


```python
megaPerturbed1 = frozen.detuningAndPinPerturbationSim(Manual_Detuning=man_detun,
                                                      Manual_Pin=man_pin,
                                                      saveSolver=True)
megaPerturbed1.plotSolitonPos()
```

Here are the peak Acav values over the course of the simulation.

```python
megaPerturbed1.plotMaxAcavAmplitudes()
```

Looks like the perturbation operations are linear! We can do them together at once, or one after the other and get the same results!
